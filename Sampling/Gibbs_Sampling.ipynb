# Gibbs Sampling

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
# generate bivariate gaussian samples
generated_samples = np.random.multivariate_normal([0,0], [[1, 0.5], [0.5,1]], 10000)
generatedx = generated_samples[:,0]
generatedy = generated_samples[:,-1]
plt.scatter(generatedx, generatedy, s = 5)
```

## Gibbs Sampling

```python
xsamples = [1]
ysamples = [-1]
N = 10000

for _ in range(N):
    prev_y = ysamples[-1]
    x_can = np.random.normal(prev_y/2, np.sqrt(3/4))
    y_can = np.random.normal(x_can/2, np.sqrt(3/4))
    xsamples.append(x_can)
    ysamples.append(y_can)
```

```python
plt.scatter(xsamples, ysamples, s = 5)
```

```python
# check the correlation between previous samples and current samples

# in generated samples

print(f"Correlation of x = {pearsonr(list(generatedx)[:-1] , list(generatedx)[1:] )[0]}")
print(f"Correlation of y = {pearsonr(list(generatedy)[:-1] , list(generatedy)[1:] )[0]}")
```

```python
# in gibbs samples

print(f"Correlation of x = {pearsonr(xsamples[:-1] , xsamples[1:] )[0]}")
print(f"Correlation of y = {pearsonr(ysamples[:-1] , ysamples[1:] )[0]}")
```
